## Cleaning the data and Listing Features with missing Values

Check for Blanks or NA's!

Note that there is a difference between NaN, '', None so be aware when looking at your data

NaN = not a number, still a float type, so think of it of empty space that can still be passed through numerical operations
'' = is a empty string type
None = is also a empty space, but in DataFrames it is considered an object which cannot be processed through optimized numerical operations
Pandas: Check for NA's in a column

```python
total = df_data.isnull().sum().sort_values(ascending=False)
percent = (df_data.isnull().sum()/df_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
# fillna with mean for the remaining columns: LotFrontage, GarageYrBlt, MasVnrArea
df_data.fillna(df_data.mean(), inplace=True)
```

```python
total = df_data.isnull().sum().sort_values(ascending=False)
percent = (df_data.isnull().sum()/df_data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head()
```

```python
# Checking For missing Values After Replacing

df_data.isnull().sum().sum()
```

```python
#Clean outliers

def clean_outliers(data, outliers):
    for col in outliers:
        data = data[data[col] < outliers[col]]
    return data

print("Before cleaning: %d" % len(df_data))
df_data = clean_outliers(df_data, outliers)
print("After cleaning: %d" % len(df_data))
```

```python
#Find the appropriate features sing heat-map 
data = df_data

plt.subplots(figsize=(20,15))
correlation_matrix = data.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
plt.show()
```

There were positive and negative correlations found in the dataset as shown in the heat-map. Correlation threshold is set to 0.6 to select the right independent variables

```python
corre_thresh = 0.6

corr = data.corr()
cols = data.columns
for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        attr1 = cols[i];
        attr2 = cols[j];
        if corr[attr1][attr2] > corre_thresh:
            print("(%s, %s): %.2f" % (attr1, attr2, corr[attr1][attr2]))
```

```python
drop = ['GarageYrBlt', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'TotRmsAbvGrd']


data = data.drop(drop, axis=1)
data
```

```python
#Split the dataset to independent variables and dependent variable
X = df_data.iloc[:,:-1].values
y = df_data.iloc[:,-1].values
```

```python
#Split the dataset to training set and testing set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,random_state = 15, shuffle=True)
```

## Model Fitting

```python
#Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

## Testing the prediction

```python
#Predict 
y_pred = linear_regressor.predict(X_test)
```

## Visualising by Scatter plot

```python
#Plot
plt.scatter(y_test, y_pred)
plt.xlabel("Real SalesPrice")
plt.ylabel("Predicted SalesPrice")
plt.show()
```
